```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
def gauss(x,mu,sigma):
    c = 1.0 / (np.sqrt(2*np.pi*sigma))
    cc = (-1 * (x-mu)**2) / (2*sigma)
    y = c*np.exp(cc)
    return y
    
```

```python
gauss(1.0,0.0,0.2)
```

```python
n = 10000.0
n_bins = 100
xmin = -5
xmax = 5
sigma = 0.5
a = np.random.normal(1,np.sqrt(sigma),int(n))

def myhist(a):
    yh,xh = np.histogram(a,n_bins,range=[-5,5])
    return yh,xh

yh,xh = myhist(a)

x = np.linspace(xmin,xmax,100)
y = [(n*(xmax-xmin)/n_bins)*gauss(i,1,sigma) for i in x]

plt.clf()
plt.plot(xh[:-1],yh)
plt.plot(x,y,"r--",linewidth=5)


```

```python
yh2,xh2 = myhist(a+np.random.permutation(a))
plt.plot(xh2[:-1],yh2)
```
